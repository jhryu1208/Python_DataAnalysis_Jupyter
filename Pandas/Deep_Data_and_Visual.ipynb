### [ 1. 데이터 분석을 위한 '위치 추정' 과 '범위 추정' ]

탐색적 데이터 분석의 STEP 3. feature 분석을 위한 기본 탐색
1. <b>위치추정</b> : 방대한 데이터의 대표값을 구해서, 해당 feature의 일종의 요약 정보를 도출
    - 보통은 <b>평균</b>을 사용한다. ( 하지만, 평균이 가장 좋은 대표값이 아닌 경우도 많다. )




2. <b>변이추정</b> : 방대한 데이터의 분포를 알아내어, 해당 feature의 일종의 요약 정보를 도출
    - <b>표준편차</b> 와 <b>분산</b>이 대표적인 분포값에 해당한다.

#### (1) 위치추정

#### 평균 (mean)
- 모든 값을 갯수로 나눈 값

#### 가중 평균 (weighted mean)
- 데이터값($x_i$) X 가중치($w_i$) 의 총 합을 다시 가중치($w_i$)의 총합으로 나눈 것
- 예: 여러 기기로부터 가져온 데이터 중, 특정 기기는 신뢰도가 떨어질 경우, 해당 기기로부터 나온 데이터에는 가중치를 낮게 줌

#### 중간값 (median) (가중 중간값도 가능)
- 데이터를 정렬한 후 중간에 위치한 값을 취함
- 평균은 특이값(outlier)에 큰 영향을 받으므로, 특이값에 큰 영향을 받지 않도록 중간값을 활용할 수 있음

#### 절사평균 (trimmed mean)
- 데이터를 <b>정렬</b>한 후, 양끝에서 일정 개수의 값들을 빼고, 남은 데이터를 기반으로 평균을 계산
- 즉, 특잇값을 평균을 구할 때 제외하는 것임


> (!!)EDA 에서는 중간값(또는 중앙값이라고도 불리움)을 평균보다 중요하게 사용 (중간값은 특잇값의 영향을 받지 않기 때문임)

#### (2) 변이추정

#### 분산(variance)
- 평균과 각 데이터간의 차를 제곱한 값들의 합을 데이터 갯수로 나눈 값


$$ \begin{align}
\sigma^2 = \dfrac{\sum_{i = 1}^{N} (x_i - \mu)^2}{N}
\end{align} $$

- $ \sigma^2 $ : 모분산 (깊게 들어가면 샘플분산은 n - 1 로 나누지만, 분산의 형태만 간략히 참고로 알아두기로 함)
- $ \mu $ : 평균
- $ N $ : 데이터 갯수
- $ x_i $ : i 번째 데이터

#### 표준편차(standard deviation)
- 분산은 수치가 너무 커서, 분산의 제곱근 (루트를 씌운 값)
- $ \sigma $ : 모표준편차 (깊게 들어가면 샘플표준편차는 샘플분산이 n 이 아닌, n - 1 로 나눈 값이므로 이를 제곱근한 값)

```python
import pandas as pd

df = pd.DataFrame({
    'A' : [1, 2, 3, 4, 5, 6],
    'B' : [1, 2, 3, 4, 5, 100]
    })

df
```

- 이때, describe() 함수를 통해서 데이터프레임의 평균, 표준편차, 중간값을 확인 할 수 있다.


    - 평균 : mean
    - 표준편차 : std
    - 중간값 : 50%  (중간값은 행이 짝수 일 경우 50%위치에서의 앞과뒤를 더한뒤 2를 나눈값이다.)

```python
df.describe()
```

### [ 2. 더 빠른 분석을 위한 데이터 분류 이해 ]
- 데이터를 분석, 시각화, 예측하는 전반적인 과정에서 데이터에 대한 큰 그림을 이해하는 것이 도움이 된다.


#### ※ 데이터는 크게 "수치형", "범주형" 데이터로 나눌 수 있다.


<b>1. 수치형 데이터</b>
- 연속(continuous) 데이터 : 특정한 범위 안에 어떤 값(정수와 <b>부동 소수점</b>)이든 가질 수 잇는 데이터
- 이산(discrete) 데이터 : 횟수와 같은 <b>정수</b>만 가질 수 있는 데이터

<b>2. 정수형 데이터</b>
- 명목(nominal) 데이터 : 카테고리, 타입, 항목등 데이터 분류를 위해 이미 정해진 값이 있는 데이터, 데이터가 가질 수 있는 값을 수준(levels) 라고 부른다.
- 이진(binary) 데이터 : True/False, 0/1과 같이 이미 정해진 두 값만 있는 데이터
- 순서형(ordinary) 데이터 : 이미 정해진 값 사이의 순서 관계가 있는 데이터 

ex1) 매우불만(1) ~ 매우만족(5)와 같이 분류된 값 사이에 순서가 있는 경우 - 순서형 데이터


ex2) 1반(1), 2반(2), 3반(3)을 나타내는 데이터 - 명목형 데이터

### [ 5.  분석 타입에 따른 그래프 종류 이해: 수치형 데이터 "분포"를 확인하기 위한 시각화 ]
- 도수분포표 (frequency table): 수치형 데이터를 구간으로 나눠서 각 구간에 속하는 데아터의 갯수를 나타내는 표
- 히스토그램 (histogram) 그래프: 도수 분포표를 시각적으로 표현한 그래프

```python
import numpy as np
import pandas as pd

df = pd.DataFrame( np.random.rand(10000,1), columns = ['A'])
df.head()
```

#### (1) iplot() 으로 histogram 타입 그래프 그려보기

```python
import chart_studio.plotly as py
import cufflinks as cf
cf.go_offline(connected = True)

df.iplot(kind = 'histogram')
```

```python
# 구간을 10개로 균등하게 나누기
df.iplot(kind = 'histogram', bins = 10)
```

#### (2) plotly.graph_objects 로 그려보기

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Histogram(
        x = df['A'],  # 여기서는 y가 필요없고, x만 쓰인다. 
        name = 'A',
        xbins = dict(
            start = 0,
            end = 1.0,
            size = 0.05
        ),
        marker_color = '#F50057'
    )
) 

fig.update_layout({
    
    'title' : 'Sampled Results',
    
    'xaxis' : {
        'title' : 'Value'},
    
    'yaxis' : {
        'title' : 'Count'},
    
    'bargap' : 0.1
    }
)


# layout은 아래처럼도 표현 할 수 있다.

# fig.update_layout(
#     title_text='Sampled Results', # title of plot
#     xaxis_title_text='Value', # xaxis label
#     yaxis_title_text='Count', # yaxis label
#     bargap=0.1, # gap between bars of adjacent location coordinates

# )


fig.show()
```
