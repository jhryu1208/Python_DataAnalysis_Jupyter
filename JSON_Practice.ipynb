### JSON 데이터 포멧
 - JSON 포멧 예시 <br>
 { "id":"01", "language": "Java", "edition": "third", "author": "Herbert Schildt" }

### EX1) json.loads() 함수로 문자열로된 json 데이터 읽기

````python
import json

data = ' { "id":"01", "language": "Java", "edition": "third", "author": "Herbert Schildt" }'
# 문자열로 된 JSON 포멧의 데이터
json_data = json.loads(data)

# key,value 출력
print(json_data.items(), type(json_data))
print()
# key 출력
print(json_data.keys(), type(json_data))
print()
# value 출력
print(json_data.values(), type(json_data))
print(json_data['id'], json_data['language'], json_data['edition'], json_data['author'])   
````

### EX2) json.dumps() 함수로 파이썬 사전 데이터를 json 문자열 데이터로 변환하기

````python
import json

data = { "id":"01", "language": "Java", "edition": "third", "author": "Herbert Schildt" }
# 사전 데이터 형식의 JSON 데이터

str_json_data = json.dumps(data)
print(str_json_data, type(str_json_data))

print()

str_json_data_applyindent = json.dumps(data, indent=2) # *** indent=n 을 이용한 들여쓰기 (n칸 제공받음)
print(str_json_data_applyindent)
````

### EX3) json.dump() 함수로 파이썬 사전 데이터 파일 쓰기 (*json.dumps()와는 다르다!!)

````python
import json

data = { "id":"01", "language": "Java", "edition": "third", "author": "Herbert Schildt" }
data['language'] = ['Java', 'C']

with open('../DataAnalysis_Ref/00_data/test.json', 'w', encoding='utf-8-sig') as json_file:
    json_string = json.dump(data, json_file, indent=2)    
    # json.dump(data, json_file, indent=2) 으로 변수없이 써도 정상적으로 파일이 작성됨
````
